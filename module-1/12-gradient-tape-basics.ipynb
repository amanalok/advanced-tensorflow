## Gradeint Tape Basics

```python
x = tf.ones(shape=(2, 2))
with tf.GradientTape() as tape:
  tape.watch(x)
  y = tf.reduce_sum(x)
  z = tf.square(y)
dz_dx = tape.gradient(z, x)
dz_dx
```

```python
x = tf.constant(3.0)
with tf.GradientTape(persistent=True) as tape:
  tape.watch(x)
  y = x * x
  z = y * y
dz_dx = tape.gradient(z, x)
print(dz_dx)
dy_dx = tape.gradient(y, x)
del tape
print(dy_dx)
```

```python
x = tf.Variable(3.0)
with tf.GradientTape(persistent=True) as tape:
  tape.watch(x)
  y = x * x
  z = y * y
dz_dx = tape.gradient(z, x)
dy_dx = tape.gradient(y, x)
del tape
print(dz_dx)
print(dy_dx)
```

```python
x = tf.Variable(3.0)
with tf.GradientTape() as tape_2:
  with tf.GradientTape() as tape_1:
    y = x * x * x
  dy_dx = tape_1.gradient(y, x)
d2y_dx2 = tape_2.gradient(dy_dx, x)
print(dy_dx)
print(d2y_dx2)
```

```python
x = tf.constant(3.0)
with tf.GradientTape() as tape:
  tape.watch(x)
  y = x * x
  z = y * y
dy, dx = tape.gradient(z, [y, x])
print(dy)
print(dx)
```
